## 4. Create new columns containing the total score of each of the personality test subscales. To do this, write a function (or functions) that will calculate the total score for each of the subscales (conscientiousness, emotional stability, openness to new experience, agreeableness, extraversion), as set out in scoring. In other words, for the conscientiousness total score, all items marked as belonging to ‘conscientiousness’ should be summed.

```python
#Dropping columns
personality_scores = personality_scores.dropna(axis = 'columns')
personality_scores.head()
```

```python
#Naming the columns
column_names = dict({1:'Extraversion', 2:'Agreeableness', 3:'Conscientiousness', 4:'Emotional Stability', 5:'Intellect'})
```

```python
#Creating a function to calculate the sun for each of the subscales
import re

def total_score(entry,column_names):
    numbers = 0
    _r = re.compile('\d+')
    for col in entry:
        num = _r.findall(col)
        if str(column_names) == num[0]:
             numbers += int (num[1])
        else:
            numbers = numbers
    return numbers

for a in column_names.keys():
    personality_scores[f'{column_names[a]}'] = personality_scores.loc[:, 'Section 5 of 6 [I am always prepared.]':'Section 5 of 6 [I worry about things.]'].apply(total_score, args = (a,), axis = 1)
```
